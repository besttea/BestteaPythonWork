# 结构力学例题


```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
l=5
q=20
x=np.linspace(0,l,50)
M=q/2*(l*x-x**2)
V=q*(l/2-x)
print(M)
print(V)
```

```python
plt.figure(figsize=(10,4))
plt.plot([0]*len(x),color="k")
plt.plot(-M)
plt.ylim(top=(q*1/2)+50)
plt.plot(V)
plt.show()

```

```python
from scipy.optimize import fsolve,root
l=5
q=20
x=np.linspace(0,10,50)
y1=-x*x+16*x
y2=-0.5*x*x+12*x+6
def poly(xs):
    x=xs
    err=-x*x+16*x+0.5*x*x-12*x-6
    return err   
func=lambda x : -x*x+16*x+0.5*x*x-12*x-6
result=fsolve(func,6)
result2=root(poly,x0=[0],args=(0,10))

print(result2)
  
plt.plot(y1)
plt.plot(y2)
plt.plot(result,"o")
plt.show
```
